```python
import sys
sys.path.insert(0, '../src')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from data.loader import load_raw_data
from data.preprocessor import ChurnPreprocessor, prepare_data, TARGET

df = load_raw_data()
print(f'Raw shape: {df.shape}')
df.head(3)
```

## 1. Missing Values & Duplicates

```python
print('Duplicates:', df.duplicated().sum())
print('\nMissing values:')
missing = df.isnull().sum()
print(missing[missing > 0] if missing.any() else 'None ✅')
```

## 2. Preprocessing Pipeline

```python
X_train, X_test, y_train, y_test, preprocessor = prepare_data(df)
print(f'\nFeatures generated: {X_train.shape[1]}')
print('\nFirst 10 features:', list(X_train.columns[:10]))
```

## 3. Class Imbalance

```python
print(f'Train churn rate: {y_train.mean():.1%}')
print(f'Test churn rate:  {y_test.mean():.1%}')

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
for ax, (split_name, y_split) in zip(axes, [('Train', y_train), ('Test', y_test)]):
    counts = y_split.value_counts()
    ax.bar(['Retained', 'Churned'], counts.values, color=['#2A9D8F', '#E63946'])
    ax.set_title(f'{split_name} Set Class Distribution', fontweight='bold')
    ax.set_ylabel('Count')
    for i, v in enumerate(counts.values):
        ax.text(i, v + 10, f'{v:,}\n({v/len(y_split):.1%})', ha='center', fontsize=9)
plt.suptitle('Class Balance After Stratified Split', fontweight='bold')
plt.tight_layout()
plt.show()
```
